```python
#loading the data
import numpy as np
import pandas as pd
df = pd.read_excel('embeddingsdata.xlsx')
df 
```

```python
df.shape #no. of rows and columns in dataframe
```

```python
df['Label'].value_counts() #labels
```

```python
vec1 = np.array(df.loc[df['Label'] == 0]) #label 0 all features
vec1
```

```python
vec2 = np.array(df.loc[df['Label'] == 1]) #label 1 all features
vec2
```

```python
mean_vec1 = np.mean(vec1, axis = 0) # mean for all features of label 0
mean_vec1
```

```python
mean_vec2 = np.mean(vec2,axis = 0) # mean for all features of label 1
mean_vec2
```

```python
spread_vec1 = np.std(vec1,axis = 0) # standard deviation for all features of label 0
spread_vec1
```

```python
spread_vec2 = np.std(vec2,axis = 0) # standard deviation for all features of label 1
spread_vec2
```

```python
interclass_distance = np.linalg.norm(mean_vec1 - mean_vec2) #interclass distance between label 0 and label1
interclass_distance
```
